                                                          Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
BSN=BS1[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']]
# Loop through each column in the DataFrame BS1
for i in BSN:
    plt.figure()  # Create a new figure for each histogram
    plt.hist(BSN[i])  # Create a histogram for the current column
    plt.title(f'Histogram of {i}')  # Set the title of the histogram
    plt.xlabel(i)  # Label the x-axis
    plt.ylabel('Frequency')  # Label the y-axis
    plt.show()  # Display the histogram
```

```python
for i in BSN:
    Q1 = BSN[i].quantile(0.25)
    Q3 = BSN[i].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = BSN[(BSN[i]< lower_bound) | (BSN[i] > upper_bound)]
    print(f"Column: {i}, Lower bound: {lower_bound}, Upper bound: {upper_bound},\n Outliers: {outliers[i].values},\n Oulier_count: {outliers[i].count()} ")
    print("-" * 10)
```

Conclusion:

From the above histogram we conclude that apart from "Discount Rate (%)" every field is Right Skewed and there is presence of outliers in all the fields

```python
# we can also find skewness as below 
from scipy.stats import skew
skewness_value = skew(BSN)
print(f"{BSN.columns},skewness:{skewness_value}")
```

```python
# Standardization of Numerical Variables
# ●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.
# ●	Steps:
# 	Explain the concept of standardization (z-score normalization).
# 	Standardize the numerical columns using the formula: z=x-mu/sigma
# 	Show before and after comparisons of the data distributions.
```

Step-1

```python
# Standardization : Data standardization is the process of transforming data from various sources into a consistent format so that
#                   it can be compared and interpreted meaningfully.
```

Step-2

```python
BS1.info()
```

```python
BSN=BS1[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']]
BSN.head()
```

Step-3

```python
BS1.describe()
```

```python
# Standardize the columns in BSN using Z-score formula (x-mean/std)
BSN_standardized = BSN.copy()
for i in BSN.columns:
# Calculate mean (mu) and standard deviation (sigma) for each column
    mu = BSN[i].mean()
    sigma = BSN[i].std()
    print(f"{i} mean:{mu}, {i} sigma:{sigma}\n")
    
# Apply Z-score formula: z = (x - mu) / sigma
    BSN_standardized[i] = (BSN[i] - mu) / sigma
    
# Display the standardized data

print(BSN.head())
print(BSN_standardized.head())
```
